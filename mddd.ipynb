```python
# 3일 연속 음봉 거르는 함수 만들기 / start_date 기준
def find_stocks_with_negative_returns(start_date,df='daily'): # df 디폴트를 daily로 설정
    negative_return_stocks = []
    
    # Calculate the index position of the start date
    start_date_index = df.index.get_loc(start_date)
    
    for column in df.columns:
        consecutive_negatives = 0
        for i in range(start_date_index, 0, -1):
            return_value = df[column].iloc[i]
            if return_value < 0:
                consecutive_negatives += 1
                if consecutive_negatives == 3:
                    negative_return_stocks.append(column)
                    break
            else:
                consecutive_negatives = 0
    
    return negative_return_stocks
```

```python
def find_top_3_stocks_with_highest_mdd(start_date, df='daily'):
    # Find stocks with negative returns for three consecutive days leading up to the start_date
    negative_return_stocks = find_stocks_with_negative_returns(df, start_date)
    
    # Calculate MDD values for each selected stock from 3 days to 1 day before the start_date
    mdd_results = {}
    
    # Calculate the index position of the start date
    start_date_index = df.index.get_loc(start_date)
    
    for stock_code in negative_return_stocks:
        mdd_stock = []
        for i in range(start_date_index - 3, start_date_index):
            # Calculate MDD for the 3-day period
            three_day_returns = df.iloc[i:i+3][stock_code]
            mdd = (1 - three_day_returns / three_day_returns.max()).max()
            mdd_stock.append(mdd)
        mdd_results[stock_code] = mdd_stock
    
    # Rank stocks by MDD and select the top 3
    top_3_stocks = []
    
    for stock_code, mdd_values in mdd_results.items():
        if len(mdd_values) > 0:
            max_mdd = max(mdd_values)
            top_3_stocks.append((stock_code, max_mdd))
    
    top_3_stocks.sort(key=lambda x: x[1], reverse=True)
    top_3_stocks = top_3_stocks[:3]
    
    return top_3_stocks
    
```

```python
def calculate_daily_returns(start_date, top_3_stocks,df='daily'):
    # Filter 'df' to include only the selected stocks and the 'start_date'
    selected_returns = df[top_3_stocks + [start_date]]
    
    # Calculate the daily returns for the selected stocks on the 'start_date'
    daily_returns = selected_returns.loc[start_date]
    
    return daily_returns
```
